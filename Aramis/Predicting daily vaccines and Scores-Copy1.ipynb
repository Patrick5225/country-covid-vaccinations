### Turning the variable values into Scores

```python
#Receives a particular column and returns a score: a number between 1 and 10, depending on the bin that that number fits
def normalized_score(df_src, column_name, apply_log=True):
    i=0
    df=df_src.copy()
    if apply_log:
        df['log']=np.log(df[column_name]+0.00001)
    else:
        df['log']=df[column_name]
    
    df['log_score']=0
    for i in range(0,10):
        j=0
        for j,v in df['log'].items():
            if df.log[j] in df['log'].value_counts(bins=10, sort=False,normalize= True).index[i]:
                df['log_score'][j] = i+1 # score of 10 for the higher numbers
        
            if df.log[j]==df.log[j] and df['log_score'][j]==0: # special exception for the highest number, for it gets excluded from the interval
                df['log_score'][j]=10
    
    return df['log_score']
```

```python
#now we create a dataframe to store the locations and the respective selected variables scores

dfD=pd.DataFrame(dfC['location'])
dfD['gdp_score']      =normalized_score(dfC, 'gdp_per_capita', apply_log=True)
dfD['HDI_score']      =normalized_score(dfC, 'HDI', apply_log=False) # HDI is already a index containing log
dfD['daily_vac_score']=normalized_score(dfC, 'pred_vac_day_pop', apply_log=True)
dfD['pop_score']      =normalized_score(dfC, 'population', apply_log=True)
dfD['distance']      =normalized_score(dfC, 'distance', apply_log=True)
```
