```python
#Here we are going to run the lazyRegressor sometimes with different states, because we have a small dataset so the R2
#statistic can vary signifcantly
i=0
models_sum=[[]]
for i in range(0,20):
    X_train1, X_test1, Y_train1, Y_test1 =train_test_split(X1,Y1,test_size=.25,random_state =i)
    regr=LazyRegressor(verbose=0,predictions=True)
    models_r,predictions_r=regr.fit(X_train1, X_test1, Y_train1, Y_test1)
    if i == 0:
        models_sum=models_r
    else:
        models_sum= models_sum + models_r
models_sum=models_sum/20
```

### Turning the variable values into Scores

```python
#Receives a particular column and returns a score: a number between 1 and 10, depending on the bin that that number fits
import math

def normalized_score(df_src, column_name, apply_log=True, higher_better=True):
    i=0
    df=df_src.copy()
    if apply_log:
        df['log']=np.log(df[column_name]+0.00001)
    else:
        df['log']=df[column_name]
    
    df['log_score']=0
    for i in range(0,10):
        for j,v in df['log'].items():
            if math.isnan(df.log[j]): continue
            if df.log[j] in df['log'].value_counts(bins=10, sort=False,normalize= True).index[i]:
                df['log_score'][j] = i+1 # score of 10 for the higher numbers
        
            if df.log[j]==df.log[j] and df['log_score'][j]==0: # special exception for the highest number, for it gets excluded from the interval
                df['log_score'][j]=10
    
    if higher_better==False: #revert the score if the variable is "the lower the better"
            df['log_score'] = -df['log_score'] +11
            df.loc[df.log_score == 11, 'log_score'] = 0
    
    
    # let's make this score have a maximum of 1:
    
    df['log_score']=df['log_score']/10
    
    # The following lines are going to show the table with the lower and higher limits for the scores
    
    print(column_name)
    
    if apply_log==True:
        for i in range(0,10):
            print(math.exp(df['log'].value_counts(bins=10, sort=False,normalize= True).index[i].left),',',math.exp(df['log'].value_counts(bins=10, sort=False,normalize= True).index[i].right),',',(i+1)/10)
    else:    
        for i in range(0,10):
            print(df['log'].value_counts(bins=10, sort=False,normalize= True).index[i].left,',',df['log'].value_counts(bins=10, sort=False,normalize= True).index[i].right,',',(i+1)/10)
    
    return df['log_score']
```

```python
df_LPI=pd.read_csv('LPI_rank_score.csv')
```

```python
df_LPI['lpi_score'].value_counts(bins=10,sort=False,normalize=True)
```

```python
dfC = dfC.merge(df_LPI,
                how='left',
                right_on='location',
                left_on='location')
```

```python
dfC[['location','lpi_score']]
```

```python
#now we create a dataframe to store the locations and the respective selected variables scores

dfD=pd.DataFrame(dfC['location'])
dfD['gdp_score']      =normalized_score(dfC, 'gdp_per_capita', apply_log=True)
dfD['HDI_score']      =normalized_score(dfC, 'HDI', apply_log=False) # HDI is already a index containing log
dfD['daily_vac_score']=normalized_score(dfC, 'pred_vac_day_pop', apply_log=True)
dfD['pop_score']      =normalized_score(dfC, 'population', apply_log=True)
dfD['distance']      =normalized_score(dfC, 'distance', apply_log=True, higher_better=False)
dfD['LPI_score']      =normalized_score(dfC, 'lpi_score', apply_log=False) # LPI doesn't show much discrepancy requiring log function
```

```python
dfD.corr()
```

```python
gdp_weight      =0.0
HDI_weight      =0.0
daily_vac_weight=0.40
pop_weight      =0.40
distance_weight =0.00
LPI_weight      =0.20
```

```python
dfD['final_score']=dfD['gdp_score']*gdp_weight + dfD['HDI_score']*HDI_weight + dfD['daily_vac_score']*daily_vac_weight + dfD['pop_score']*pop_weight +dfD['distance']*distance_weight+dfD['LPI_score']*LPI_weight
```

```python
pd.options.display.float_format = '{:,.2f}'.format
dfD.sort_values(by=['final_score'], ascending=False)
```

### Simple Sensitive Analysis

```python
#{'gdp_weight': [gdp_weight],'HDI_weight': [HDI_weight],'daily_vac_weight': [daily_vac_weight],'pop_weight': [pop_weight],'LPI_weight': [LPI_weight]}
weight = {'daily_vac_weight': [daily_vac_weight],'pop_weight': [pop_weight],'LPI_weight': [LPI_weight]}
w_df = pd.DataFrame(data=weight)
```

```python
# Here we will take turns raising the importance of a particular weight and
# reducing the importance of the others to see the most stable option 

def simple_sens(w_df,df,amplitude=0.15):
        for i in range(0,w_df.shape[1]):
            w_df = w_df - amplitude/(w_df.shape[1])
            w_df.iloc[0, i]=w_df.iloc[0, i]+amplitude
            df['final_score']=dfD['daily_vac_score']*w_df.iloc[0, 0] + dfD['pop_score']*w_df.iloc[0, 1]+dfD['LPI_score']*w_df.iloc[0, 2]
            print(w_df.columns[i])
            print(df[['location','final_score']].sort_values(by=['final_score'], ascending=False).head(10))
            print("\n")
   
```

```python
simple_sens(w_df,dfD,amplitude=0.15)
```
